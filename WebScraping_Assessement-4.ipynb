## 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:
## A) Rank
## B) Name
## C) Artist
## D) Upload date
## E) Views 

```python
driver = webdriver.Chrome()
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
Rank=[]
Name=[]
Artist=[]
Date=[]
View=[]
```

```python
try:
    for i in range(1,31):
        rank = driver.find_element(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[2]/tbody/tr[{}]/td[1]'.format(i))
        Rank.append(rank.text)
        title=driver.find_element(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[2]/tbody/tr[{}]/td[2]'.format(i))
        Name.append(title.text)
        artist=driver.find_element(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[2]/tbody/tr[{}]/td[3]'.format(i))
        Artist.append(artist.text)
        date=driver.find_element(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[2]/tbody/tr[{}]/td[5]'.format(i))
        Date.append(date.text)
        view=driver.find_element(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[2]/tbody/tr[{}]/td[4]'.format(i))
        View.append(view.text)
except NoSuchElementException as e:
    pass
```

```python
ds1=pd.DataFrame({'Rank':Rank,'Title':Name,'Artist':Artist,'Published date':Date,'Views':View})
ds1
```

```python
driver.quit()
```

```python
for j in range(0,10):
    try:
        element = driver.find_element(By.XPATH,'//*[@id="fixtures"]/div[3]/div[2]/div/button')
        driver.execute_script("arguments[0].click();", element)
    except NoSuchElementException as e:
        break
```

```python
match=[]
Series=[]
Place=[]
Date=[]
Time=[]
Team1=[]
Team2=[]
```

## 8. Details of Datasets from UCI machine learning repositories.
## Url = https://archive.ics.uci.edu/

## A) Dataset name
## B) Data type
## C) Task
## D) Attribute type
## E) No of instances
## F) No of attribute 
## G) Year 

```python
driver = webdriver.Chrome()
driver.get('https://archive.ics.uci.edu/')
```

```python
try:
    view=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
    driver.execute_script("arguments[0].click();", view)
except NoSuchElementException as e:
    view=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
    driver.execute_script("arguments[0].click();", view)
```

```python
try:
    expand=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]')
    driver.execute_script("arguments[0].click();", expand)
except NoSuchElementException as e:
    expand=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]')
    driver.execute_script("arguments[0].click();", expand)
```

```python
Name=[]
Data_type=[]
Task=[]
Att_type=[]
Instances=[]
Attribute=[]
Year=[]
```

```python
import time
```

```python
page=int(1)
while page<66: 
        for j in range(0,2):
            expand=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]')
            driver.execute_script("arguments[0].click();", expand)
        title=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
        for i in title:
            Name.append(i.text)
        for k in range(1,20,2):
            typed=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[{}]/div[2]/div/table/tbody/tr/td[1]'.format(k))
            Data_type.append(typed.text)
        des=driver.find_elements(By.XPATH,'//p[@class="truncate"]')
        for i in des:
            Task.append(i.text)
        for k in range(1,20,2):
            att_type=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[{}]/div[2]/div/table/tbody/tr/td[2]'.format(k))
            Att_type.append(att_type.text)
        for k in range(1,20,2):
            inst=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[{}]/div[1]/div[2]/div/div[3]/span'.format(k))
            Instances.append(inst.text)
        for k in range(1,20,2):
            att=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[{}]/div/div[2]/div/div[4]/span'.format(k))
            Attribute.append(att.text)
        for k in range(1,20,2):
            yr=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[{}]/div[2]/div/table/tbody/tr/td[3]'.format(k))
            Year.append(yr.text)
        nextp=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
        driver.execute_script("arguments[0].click();", nextp)
        time.sleep(3)
        page+=1        
```

```python
print(len(Name))
print(len(Data_type))
print(len(Task))
print(len(Att_type))
print(len(Instances))
print(len(Attribute))
print(len(Year))
```

```python
ds8=pd.DataFrame({'Dataset Name':Name,'Data Type':Data_type,'Task':Task,'Attribute type':Att_type,'No Of Instances':Instances,'No Of Attributes':Attribute,'Year':Year})
ds8
```
