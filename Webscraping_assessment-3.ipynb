```python
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException,ElementNotInteractableException
import warnings
warnings.filterwarnings('ignore')
import time
```

## 1.Write a python program which searches all the product under a particular product from www.amazon.in. Theproduct to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars.

```python
driver = webdriver.Chrome()
driver.get('https://www.amazon.in')
print('Enter the product:')
x = input()
designation = driver.find_element(By.XPATH,'//*[@id="twotabsearchtextbox"]')
designation.send_keys(x)
designation.send_keys(Keys.RETURN)
```

## 2.In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" an  “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. 

```python
item=[]
price=[]
delivery=[]
url=[]
```

```python
start=0
end=3
for page in range(start,end):

#Item name
    try:
        title_tags=driver.find_elements(By.XPATH,'//span[@class="a-size-medium a-color-base a-text-normal"]')
        for i in title_tags:
            title=i.text
            item.append(title)
    except NoSuchElementException as e:
        item.append('-')
    
#Price
    try:
        price_tag=driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
        for i in price_tag:
            prices=i.text
            price.append(prices)
    
    except NoSuchElementException as e:
        price.append('-')

#Expected delivery   
    try:
        exp_del=driver.find_elements(By.XPATH,'//div[@class="a-row a-size-base a-color-secondary s-align-children-center"]')
        for i in exp_del:
            deli=i.text
            delivery.append(deli)
   
    except NoSuchElementException as e:
        delivery.append('-')

#Product URL    
    try:
        urls=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
        for i in urls:
            link=i.get_attribute('href')
            url.append(link)
    except NoSuchElementException as e:
        url.append('-')
    
    
    next_but=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    next_but.click()
    time.sleep(3)

    
driver.quit()
```

```python
ds=pd.DataFrame({'Title':item,'Price':price,'Delivery':delivery,'URL':url})
ds
```

```python
ds.to_csv('1.web_scrapping.csv', index=True) 
```

## 3.Write a python program to access the search bar and search button on images.google.com and scrape 10 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.

```python
items=['fruits', 'cars','Machine Learning','Guitar','Cakes']

img=[]
for item in items:
    driver = webdriver.Chrome()
    driver.get('https://images.google.com/')
    designation = driver.find_element(By.CSS_SELECTOR,"#APjFqb")
    designation.send_keys(item)
    designation.send_keys(Keys.RETURN)
    for i in range(1,11):
        imgs = driver.find_element(By.XPATH,'/html/body/div[2]/c-wiz/div[3]/div[1]/div/div/div/div/div[1]/div[1]/span/div[1]/div[1]/div[{}]/a[1]/div[1]'.format(i))
        image_element = imgs.find_element(By.TAG_NAME,"img")
        img.append(image_element.get_attribute("src"))
    print('number of urls in the img list:',len(img))
    driver.quit()
```

## 4.Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV

```python
driver = webdriver.Chrome()
driver.get('http://www.flipkart.com/')
try:
    cancel=driver.find_element(By.XPATH,'/html/body/div[2]/div/div/button')
    cancel.click()
except NoSuchElementException as e:
    pass
print('Enter the Mobile brand:')
x = input()
try:
    designation = driver.find_element(By.CLASS_NAME,'_3OO5Xc')
    designation.send_keys(x)
    designation.send_keys(Keys.RETURN)
except ElementNotInteractableException as e:
    designation = driver.find_element(By.XPATH,'//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
    designation.send_keys(x)
    designation.send_keys(Keys.RETURN)
```

```python
url=[]
item=driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
for i in item:
        link=i.get_attribute('href')
        url.append(link)
```

```python
Brand=[]
name=[]
RAM=[]
clr=[]
ROM=[]
pri_cam=[]
sec_cam=[]
price=[]
battery=[]
display=[]
for i in range(0,len(url)):
#Brand
    Brand.append(x)
    driver.get(url[i])
    
#Model name
    try:
        element_name=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[2]/div/div[1]/h1/span')
        ele_name=element_name.text
        name.append(ele_name)
        
    except NoSuchElementException as e:
        name.append('-')
        
#RAM     
    try:
        element_ram=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[2]/div/div[1]/h1/span')
        ele_ram=element_ram.text
        RAM.append(ele_ram)
        
    except NoSuchElementException as e:
        RAM.append('-')

#Color
    try:
        element_clr=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[1]/table/tbody/tr[4]/td[2]/ul/li')
        ele_clr=element_clr.text
        clr.append(ele_clr)
        
    except NoSuchElementException as e:
        clr.append('-')

#ROM
    try:
        element_rom=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[4]/table/tbody/tr[1]/td[2]/ul/li')
        ele_rom=element_rom.get_attribute("textContent")
        ROM.append(ele_rom)
        
    except NoSuchElementException as e:
        ROM.append('-')
    

#Primary camera
    try:
        element_pcam=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[5]/table/tbody/tr[2]/td[2]/ul/li')
        ele_pcam=element_pcam.get_attribute("textContent")
        pri_cam.append(ele_pcam)
        
    except NoSuchElementException as e:
        pri_cam.append('-')
        
#seocndary camera
    try:

        element_scam=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[5]/table/tbody/tr[5]/td[2]/ul/li')
        ele_scam=element_scam.get_attribute("textContent")
        sec_cam.append(ele_scam)
    
    except NoSuchElementException as e:
        sec_cam.append('-')

#Price       
    try: 
        element_price=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[2]/div/div[4]/div[1]/div/div[1]')
        ele_price=element_price.get_attribute("textContent")
        price.append(ele_price)
    
    except NoSuchElementException as e:
        price.append('-')

#Display size
    try:
        element_dis=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[2]/table/tbody/tr[1]/td[2]/ul/li')
        ele_dis=element_dis.get_attribute("textContent")
        display.append(ele_dis)
        
    except NoSuchElementException as e:
        display.append('-')
        
#Battery
    try:
        element_bt=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[10]/table/tbody/tr[1]/td[2]/ul/li')
        ele_bt=element_bt.get_attribute("textContent")
        battery.append(ele_bt)
    except NoSuchElementException as e:
        battery.append('-')


driver.quit()
    
```

```python
print(len(Brand),len(name),len(RAM),len(clr),len(ROM),len(pri_cam),len(sec_cam),len(price),len(display),len(battery),len(url))
```

```python
ds1=pd.DataFrame({'Brand':Brand,'Model Name':name,'RAM':RAM,'Color':clr,'ROM':ROM,'Pri_Cam':pri_cam,'Sec_cam':sec_cam,'Price':price,'Dis_size':display,'Battery':battery,'URL':url})
ds1
```

```python
ds1.to_csv('4.web_scrapping.csv', index=False) 
```

## 5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps. 

```python
driver = webdriver.Chrome()
driver.get('https://maps.google.com/')
print('Enter the location:')
x = input()
designation = driver.find_element(By.XPATH,'//input[@id="searchboxinput"]')
designation.send_keys(x)
designation.send_keys(Keys.RETURN)
current_url = driver.current_url
coordinates = current_url.split("/@")[1].split(",")
latitude, longitude = coordinates[0], coordinates[1]
print("Latitude:", latitude)
print("Longitude:", longitude)
driver.quit()
```
